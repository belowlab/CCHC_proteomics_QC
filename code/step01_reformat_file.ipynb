## 2. Reformat and reomve control samples
1. Control samples have duplicate proteins. Need to ignore them
   * SC1-3: sample controls
   * PC1-5: plate controls
   * NC1-2: negative controls
2. Control proteins

```python
print(df_data.columns)
```

```python
# Find control samples and control proteins
print('# Control samples')
print(df_data['SampleType'].value_counts())
print('\n# Control proteins')
print(df_data['AssayType'].value_counts())

mask_sample = (df_data['SampleType']=='SAMPLE') & (df_data['AssayType']=='assay')
df_data[mask_sample].shape
```

```python
# Check names of control samples and control proteins
mask_ctrl_samples = df_data['SampleType']!='SAMPLE'
print('# Control samples names')
print(df_data[mask_ctrl_samples]['SampleID'].unique())

mask_ctrl_protein = df_data['AssayType']!='assay'
print('\n# Control proteins names')
print(df_data[mask_ctrl_protein]['Assay'].unique())
```

```python
# Name of control samples: controls = ['NC1', 'NC2', 'PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'SC1', 'SC2', 'SC3']
output_path = '/data100t1/home/wanying/CCHC/proteomics/20241014_data_cleaning/output'
for col_name in ['NPX', 'PCNormalizedNPX', 'ExtNPX']:
    print(f'# Process {col_name}')
    protine_id = 'OlinkID'
    lst_dfs = []
    for labid, df in df_data[mask_sample].groupby('SampleID'):
        df = df[[protine_id, col_name]].set_index(keys=protine_id).T.reset_index().rename(columns={'index':'LABID'})
        df['LABID'] = labid
        lst_dfs.append(df)
    df_merged = pd.concat(lst_dfs)
    print('# -', df_merged.shape)
    output_fn = f'{output_path}/20240919_CCHC_proteomics.batch2.OlinkID.control_reomved.{col_name}.csv'
    df_merged.to_csv(output_fn, index=False)
```

## 3. Get plate number
Save plate number for future reference

```python
print(df_data.columns)
df_data.head()
```

```python
df_data['Plate_number'] = df_data['PlateID'].apply(lambda x: x.split('Piper_Joe_Rack_')[-1].split('_')[0])
cols_to_save = ['SampleID', 'PlateID', 'WellID', 'Plate_number']
output_fn = f'{output_path}/plate_number.all.csv'
df_plate_number = df_data[cols_to_save].drop_duplicates(subset='SampleID').rename(columns={'SampleID':'LABID'})
df_plate_number.to_csv(output_fn, index=False)
```

```python
df_plate_number[df_plate_number['Plate_number']=='1']
# Get GLP samples
lst_glp_samples = []
for sample in df_plate_number['LABID']:
    if 'GLP' in sample:
        lst_glp_samples.append(sample)
print('# N GLP samples (4 visits):', len(lst_glp_samples))
```

```python
# Get details of the GLP samples
df_glp_samples = pd.DataFrame({'ProteomicsID':lst_glp_samples})
df_glp_samples['SampleID'] = df_glp_samples['ProteomicsID'].apply(lambda x: x.split(' ')[0])
df_glp_samples['Time'] = df_glp_samples['ProteomicsID'].apply(lambda x: x.split(' ')[1])
df_glp_samples['Visit'] = df_glp_samples['ProteomicsID'].apply(lambda x: x.split(' ')[-1])
df_glp_samples = df_glp_samples.merge(df_plate_number.rename(columns={'LABID':'ProteomicsID'}), on='ProteomicsID')
display(df_glp_samples.head())
print(df_glp_samples.shape)
output_fn = f'{output_path}/plate_number.glp_only.csv'
df_glp_samples.to_csv(output_fn, index=False)
```

```python
# For GLP samples, we want time points from the same visit (V2 or V3) to be on the same plate
# Find pairs that do not follow this rule
lst_dfs = []
for ProteomicsID, df in df_glp_samples.groupby(by='SampleID'):
    mask_v2 = df['Visit']=='V2'
    mask_v3 = df['Visit']=='V3'
    if len(df[mask_v2]['Plate_number'].unique())>1:
        lst_dfs.append(df[mask_v2])
    if len(df[mask_v3]['Plate_number'].unique())>1:
        lst_dfs.append(df[mask_v3])
df_glp_merged = pd.concat(lst_dfs)
```
